```python
import re
```

```python
class RulebasedBot:
    negative_response = ("no","nope","no thanks","not","certainly not","not really","not at all","ofcourse not","absolutely not","nay","nix","sorry")
    exit_commands = ("Thank you","thank you","Goodbye","bye","exit","quit","pause", "Alright, thank you","Alright thank you")
    
    def greet(customer):
        customer.name = input("Hi, Welcome to Healthy-Eats Kiosk! What is your name?")
        assistance = input(f"Hi {customer.name}, thank you for considering Healthy-Eats. We have plenty of healthy takeaway options, how may we assist you today?")
        if assistance in customer.negative_response:
            print("Alright, have a great day!")
            return
        print(customer.chat(assistance))
        
    def exit(customer,bot):
        for command in customer.exit_commands:
            if command in bot:
                print("Thank you for considering Healthy_Eats, enjoy the rest of your day!")
                return True
        return False
            
    def chat(customer,bot):
        while True:
            print(customer.match_reply(bot))
            bot = input("")
            if customer.exit(bot):
                break
    
    
    def __init__(customer):
        customer.support_responses = {
            'ask_about_sandwich': r'.*\b(sandwich|sandwiches|wrap|wraps)\b.*',
            'ask_about_smoothie': r'.*\b(smoothie|smoothies|shake|shakes)\b.*',
            'ask_about_fruitbowl': r'.*\b(fruit|fruits|fruit bowl|fruit bowls)\b.*',
            'ask_about_salad': r'.*\b(salad|salads|green salad|garden salad)\b.*',
            'ask_about_yogurt': r'.*\b(yogurt|yogurts|yoghurt|yoghurts)\b.*'
        }
    def match_reply(customer,bot):
        for intent, regex_pattern in customer.support_responses.items():
            found_match = re.search(regex_pattern, bot, re.IGNORECASE)
            if found_match and intent =='ask_about_sandwich':
                return customer.ask_about_sandwich()
            elif found_match and intent =='ask_about_smoothie':
                return customer.ask_about_smoothie()
            elif found_match and intent =='ask_about_fruitbowl':
                return customer.ask_about_fruit()
            elif found_match and intent =='ask_about_salad':
                return customer.ask_about_salad()
            elif found_match and intent =='ask_about_yogurt':
                return customer.ask_about_yogurt()
        return customer.no_match_intent()
            
    def ask_about_sandwich(customer):
        response =("We have some delectable options to choose from, they include (1).Chicken-mayo, (2).Grilled cheese, (3).Tuna, (4). Roast beef, (5). Veggie (Vegan option), (6). Ham and Cheese, (7). Turkey and Swiss")
        return(response)
    
    def ask_about_smoothie(customer):
        response =("We have the best smoothies in town. Options include (1).Strawberry-banana, (2).Mango-pineapple, (3).Blueberry-lemon, (4). Banana-spinach, (5). Strawberry-almond, (6). Mango-cashew")
        return(response)
    
    def ask_about_fruitbowl(customer):
        response =("We have two options for fruitbowls (1).Strawberries, Mango, Kiwi, Pawpaw (2). Blueberries, Pineapple, Grapes, Watermelon")
        return(response)
        
    def ask_about_salad(customer):
        response =("Salads are our speciality. Options include (1).Greek salad, (2).Mediterranean-chickpea salad, (3).Avocado salad (4). Easy Pasta salad, (5). Taco salad, (6) Venduras y Pollo salad and (7). The Classic Chef Salad")
        return(response)
        
    def ask_about_yogurt(customer):
        response =("Hmm, in the mood for something sweet! Our yogurt flavours include (1).Strawberry, (2).Vanilla, (3).Mango, (4). Peach, (5).Salted Caramel, (6).Chocolate Fudge, (7). Plain and Fruits")
        return(response)
        
    def no_match_intent(customer):
        response = ("I am sorry, I didn't quite understand that. Can you please rephrase?")
        return(response)

bot = RulebasedBot()
bot.greet()
```
